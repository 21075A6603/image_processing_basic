```python
black=Image.open("/content/image data/1000_F_51120130_8LO6N135eNNLstBXPVXj2O9tdRUjJ6VI.jpg")
plt.imshow(black)
plt.show()
im_nt = black.point(lambda x: 255 - x)
plt.imshow(im_nt)
plt.show()
```

```python
black2=Image.open("/content/image data/pexels-photo-2098428.jpeg")
plt.imshow(black2)
```

```python
im_nt2 = black2.point(lambda x: 255 - x)
plt.imshow(im_nt2)
plt.title("nageated picture")
plt.show()
```

```python
#log transform:
im_lt=black.point(lambda x: 255*np.log(1+x/255))
plt.title("log transform applied")
plt.imshow(im_lt)
plt.show()
```

```python
#power law transform:
im_pl=black2.point(lambda x: 255*(x/255)**0.6)
plt.imshow(im_pl)
plt.show()
```
